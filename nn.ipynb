### Приведем загруженные датасеты к нужному виду

```python
X_train = normalize(np.array([np.ravel(x) for x in x_train]))
X_test = normalize(np.array([np.ravel(x) for x in x_test]))
Y_train = np.array([one_hot(np.array(y, dtype=int), 10) for y in y_train], dtype=int)
Y_test = np.array([one_hot(np.array(y, dtype=int), 10) for y in y_test], dtype=int)
```

### Каждая строка 'X' - вектор из яркостей пикселей

```python
pd.DataFrame(X_train)
```

### Каждая строка 'y' - категориальная переменная: название столбца с единицей это класс соответствующей картинки

```python
pd.DataFrame(Y_train)
```

```python
input_size = X_train.shape[1]
output_size = Y_train.shape[1]
input_size, output_size
```

### Создаем объект нейрнонной сети
Задаем ей слои с необходимым количеством входов и выходов, а также  learning rate - коэффициент скорости обучения (шаг градиентного спуска)
По умолчанию в каждом слое акцивации в качестве функциии активации берется ReLU

```python
network = NN()
network.add_layer(Dense(input_size, 100, learning_rate = 0.05))
network.add_layer(Activation())
network.add_layer(Dense(100, 200, learning_rate = 0.05))
network.add_layer(Activation())
network.add_layer(Dense(200, output_size))
```

Задаем количество эпох (n_epochs) и размер части, на которую мы делим датасет во время обучения (batch_size)

```python
train_log_relu = network.train(X_train, Y_train, n_epochs = 64, batch_size = 64)
```

### Проверяем точность модели на тестовой выборке

```python
test_corrects = len(list(filter(lambda x: x == True, network.predict(X_test) ==  Y_test.argmax(axis=-1))))
test_all = len(X_test)
test_accuracy_relu = test_corrects/test_all
f"Test accuracy = {test_corrects}/{test_all} = {test_accuracy_relu}"
```

Создадим такую же нейросеть

```python
network = NN()
network.add_layer(Dense(input_size, 100, learning_rate = 0.05))
network.add_layer(Activation())
network.add_layer(Dense(100, 200, learning_rate = 0.05))
network.add_layer(Activation())
network.add_layer(Dense(200, output_size))
```

Но на этот раз зададим ее слоям, в качестве функции активации, сигмоиду

```python
network.layers[1].set_activation(name='sigmoid')
network.layers[3].set_activation(name='sigmoid')
```

Обучим

```python
train_log_sigmoid = network.train(X_train, Y_train, n_epochs = 64, batch_size = 64)
```

Точность на тестовой выборке

```python
test_corrects = len(list(filter(lambda x: x == True, network.predict(X_test) ==  Y_test.argmax(axis=-1))))
test_all = len(X_test)
test_accuracy_sigmoid = test_corrects/test_all
f"Test accuracy = {test_corrects}/{test_all} = {test_accuracy_sigmoid}"
```

### Теперь можем сравнить процессы обучения нейросети с разными функциями активации

### Эволюция точности на обучающей выборке в зависимости от эпохи

```python
plt.plot(train_log_relu['epoch'], train_log_relu['accuracy'])
plt.plot(train_log_sigmoid['epoch'], train_log_sigmoid['accuracy'])
plt.grid()
```

### Эволюция функции ошибок в зависимости от эпохи

```python
plt.plot(train_log_relu['epoch'], train_log_relu['loss'])
plt.plot(train_log_sigmoid['epoch'], train_log_sigmoid['loss'])
plt.grid()
```

### Теперь обучим нейросеть на большом количестве эпох, это должно повлечь за собой переобучение

```python
network = NN()
network.add_layer(Dense(input_size, 100, learning_rate = 0.05))
network.add_layer(Activation())
network.add_layer(Dense(100, 200, learning_rate = 0.05))
network.add_layer(Activation())
network.add_layer(Dense(200, output_size))
train_log_overfit = network.train(X_train, Y_train, n_epochs = 128, batch_size = 64)
```

Эволюция точности переобученной сети

```python
plt.plot(train_log_overfit['epoch'], train_log_overfit['accuracy'])
plt.grid()
```

### Сравнение эволюции функции потерь сети с 64 и 128 эпохами

```python
plt.plot(train_log_relu['epoch'][60:], train_log_relu['loss'][60:])
plt.plot(train_log_overfit['epoch'][60:], train_log_overfit['loss'][60:])
plt.grid()
```

```python
test_corrects = len(list(filter(lambda x: x == True, network.predict(X_test) ==  Y_test.argmax(axis=-1))))
test_all = len(X_test)
test_accuracy_overfit = test_corrects/test_all
f"Test accuracy = {test_corrects}/{test_all} = {test_accuracy_overfit}"
```

### Сравнение всех результатов на тестовой выборке

```python
plt.bar(['sigmoid\n (64 epoch)', 'ReLU\n (64 epoch)', 'ReLU\n (128 epoch - overfit)'], [test_accuracy_sigmoid - 0.97, test_accuracy_relu - 0.97, test_accuracy_overfit - 0.97])
```

Как видно из диаграммы, излишнее количество эпох действительно дает отрицательный результат
